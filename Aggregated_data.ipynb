```python
# Import necessary libraries
import pandas as pd
import glob
```

```python
# Folder where the csv files are located
folder_path = r'C:\Users\Ernie\Documents\GitHub\Citi Bike Viz\Resources\\'
```

```python
# Find all csv files in the folder
csv_files = glob.glob(folder_path + "*.csv")
```

```python
# List to store DataFrames
dfs = []

# Loop through each file and aggregate 100 rows
for file in csv_files:
    df=pd.read_csv(file)
    df_sample = df.sample(100)
    dfs.append(df_sample)
```

```python
# Concatenate all DataFrames into one
final_df = pd.concat(dfs, ignore_index=True)
final_df.head(10)
```

```python
# Save the aggregated data to a new CSV file
final_df.to_csv(r'C:\Users\Ernie\Documents\GitHub\Citi_Bike_Viz\Resources\aggregated_sample.csv', index=False)
```

Split Date and Time

```python
# Load the csv file
df_aggregated_sample = pd.read_csv(r'C:\Users\Ernie\Documents\GitHub\Citi_Bike_Viz\Resources\aggregated_sample.csv')
```

```python
# Convert 'started_at' and 'ended_at' columns to datetime format
df_aggregated_sample['started_at']= pd.to_datetime(df_aggregated_sample['started_at'])
df_aggregated_sample['ended_at'] = pd.to_datetime(df_aggregated_sample['ended_at'])
```

```python
# Extract the date and time parts
df_aggregated_sample['start_date'] = df_aggregated_sample ['started_at'].dt.date
df_aggregated_sample['start_time'] =df_aggregated_sample ['started_at'].dt.time
df_aggregated_sample['end_date'] = df_aggregated_sample['ended_at'].dt.date
df_aggregated_sample['end_time'] = df_aggregated_sample['ended_at'].dt.time
```

```python
# Save the updated DataFrame back to a new CSV file
df_aggregated_sample.to_csv(r'C:\Users\Ernie\Documents\GitHub\Citi_Bike_Viz\Resources\updated_file.csv', index=False)
```
